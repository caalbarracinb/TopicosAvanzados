# Cáncer de seno

En este cuaderno se documenta la construcción de una red neuronal de clasificación binaria para predecir la prevalencia del cáncer de seno a partir de la base de datos [Breast Cancer Wisconsin (Diagnostic)](https://www.kaggle.com/uciml/breast-cancer-wisconsin-data/code).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, plot_confusion_matrix,accuracy_score
from sklearn.svm import SVC
from time import sleep
import sys
%matplotlib inline
```

### 1. Lectura, limpieza y exploración de los datos

```python
df = pd.read_csv('data_cancer.csv')
```

```python
df.head()
```

Las columnas 'id' y 'Unnamed:32' no son relevantes para el problema, por lo cual se procede a descartarlas. 

```python
df = df.drop(['id','Unnamed: 32'], axis = 1)
```

```python
df.head()
```

```python
df.describe()
```

### 2. Separación entre train-test y transformación de las variables

Para la red de clasificación necesitamos que la variable a explicar, en este caso 'diagnosis', sea de caracter binario. Al momento de leer los datos, se puede apreciar que es una variable factor con dos valores M y B, que corresponden a la asignación de 'Maligno' y 'Benigno', respectivamente. Adicionalmente, para la construcción de la red, es aconsejable utilizar alguna transformación a las variables con el fin de reducir el rango de variación de estas. En este caso, se empleará una estandarización para que todas correspondan a variables con media 0 y varianza 1. 

```python
df['diagnosis'] = pd.get_dummies(df['diagnosis'], drop_first = True)
```

```python
X = df.drop('diagnosis',axis=1)
X.head()
```

```python
Y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=14)
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
```

### 3. Construcción y entrenamiento de la red

Una vez separados los datos entre el set de entrenamiento y el set de prueba y ya estandarizadas las variables, se procede a la construcción de la red de clasificación binaria. La estructura de la red es la siguiente: como capa de entrada tenemos las 30 características obtenidas a través de las imágenes diagnósticas, tiene 2 capas ocultas, cada una con 5 neuronas y una capa de salida con una función de activación sigmoide que acota las predicciones de la capa anterior entre 0 y 1 para finalmente clasificarlas a partir de un umbral de 0.5. 

```python
red = tf.keras.models.Sequential()
red.add(tf.keras.layers.Dense(5, activation = 'relu'))
red.add(tf.keras.layers.Dense(5, activation = 'relu'))
red.add(tf.keras.layers.Dense(1,activation = 'sigmoid'))
```

```python
red.compile(optimizer = 'adam', loss = 'mse', metrics = ['binary_accuracy'])
```

```python
red.fit(X_train, y_train, epochs=100, verbose = 2)
```

### 4. Desempeño del modelo

A continuación se presenta algunos hallazgos sobre la bondad de ajuste del modelo. En primer lugar la matriz de confusión nos permite observar la precisión del modelo al clasificar los casos de cáncer entre benigno y maligno respecto a los valores reales. La matriz se interpreta de la siguiente manera: para las filas y columnas que coinciden (por ejemplo, fila Beningo, columna Benigno), el valor representado significa la probabilidad de que la red clasificara bien esos registros; cuando la fila y la columna tienen etiquetas distintas, estamos hablando de falsos positivos o falsos negativos, es decir, el modelo se equivocó en la clasificación. En este caso vemos que la probabilidad de que la red clasifique adecuadamente a los pacientes sin cancer es del 98.63%, mientras que al 1.37% restante les asigna cáncer cuando estos en realidad no tienen. Por otra parte, la probabilidad de que la red identifique un caso maligno, y este en verdad sea maligno, es del 95.12%, dejando un 4.88% de probabilidad de que a alguien con cáncer la red lo clasifiqué como sin cáncer. 

```python
clf = SVC(random_state=14)
clf.fit(X_train, y_train)
```

```python
cm = plot_confusion_matrix(clf,X_test,y_test,colorbar=False,normalize='true',cmap='Reds_r',values_format='.2%',display_labels=['Benigno','Maligno'])
cm.ax_.set_xlabel('Predicción del modelo')
cm.ax_.set_ylabel('Valor real')
plt.show()
```

La precisión (accuracy) es una medida que permite evaluar el desempeño global del modelo al momento de clasificar sin tener en cuenta los casos particulares. La red construída en este cuaderno tiene una precisión del 97.36%.

```python
y_true = y_test
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_true,y_pred)
```
